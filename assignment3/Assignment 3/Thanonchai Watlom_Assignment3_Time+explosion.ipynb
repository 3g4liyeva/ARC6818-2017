```python
import turtle
import datetime
import time

#In this assignment I try to use the turtle graphic to represent the reading time by using the present time from computer.
#In this case, I use the yellow, orange, and red turtles represent second. Arter 60 seconds will start over and over 
#again to represent minute.

#Set the color and shape of the turtles
hrT = turtle.Turtle()
t=turtle.Turtle()
t.shape('turtle')
t.color('#E6EA69')
t.stamp()
t1=turtle.Turtle()
t1.shape('turtle')
t1.color('#F3B020')
t2=turtle.Turtle()
t2.shape('turtle')
t2.color('#F36A20')

move=30
x=0
y=0

#Set the present time for the blue turtle
currentTime=datetime.datetime.now().time()
hrT.penup()
hrT.goto(250,250)
hrT.setheading(90)
hrT.shapesize(5) 
hrT.shape('turtle')
hrT.color('#579CA4')
hrT.rt((currentTime.hour*30)-(90-hrT.heading()))
prevHour=currentTime.hour

#Then I try to adapt the fuction that I have learned in the class for this and adapt the function 
#from the example code to make the reading time for turtle graphic in this part

for i in range(0,100):
    currentTime=datetime.datetime.now().time()
    currentHour=currentTime.hour
    if (currentHour != prevHour):
        hrT.rt((currentTime.hour*30)-(90-hrT.heading()))
    prevHour=currentTime.hour


#In this part, I used while true to represent the loop of time that will never stop. The turtles will move
#in 360degress where the distance from each are 6degrees apart from each other.
while True:
    for i in range(60) :
        t.pencolor('#E6EA69')
        t.penup()
        t.forward(20)
        t.pendown()
        t.forward(10)
        t.penup()
        t.forward(15)
        t.stamp()
        t.home()
        t.right(move)
        move += 6
        x += 1
    if x == 61:
        x =0
    elif x == 60:
        for i in range(60) :
                t1.pencolor('#F3B020')
                t1.penup()
                t1.forward(60)
                t1.pendown()
                t1.forward(50)
                t1.penup()
                t1.forward(55)
                t1.stamp()
                t1.home()
                t1.right(move)
                move += 6
                #t.clear()   
                y += 1
                       
   
                if y == 61:
                    y = 0
                elif y == 60:
                    for i in range(60) :
                        t2.pencolor('#F36A20')
                        t2.penup()
                        t2.forward(100)
                        t2.pendown()
                        t2.forward(90)
                        t2.penup()
                        t2.forward(95)
                        t2.stamp()
                        t2.home()
                        t2.right(move)
                        move += 6
                        #t1.clear()   
```
