### This function is to output the weather of the localtime. There are four seasons which represented by four tree drawings. Use datetime function to find out the month now and then output will give a turtle drawing of the weather corresponding to the month now.

```python
import turtle




def spring_tree(branchLen,t):
    
    b=branchLen
    
    if b > 75:
        t.color('brown')
        t.pensize(5)
        t.forward(branchLen)
        t.right(20)
        spring_tree(branchLen-15,t)
        t.left(40)
        spring_tree(branchLen-15,t)
        t.right(20)
        t.pu()
        t.backward(branchLen)
        
        
    if b > 60 and b < 75:
        t.color('brown')
        t.pd()
        t.pensize(4)
        t.forward(branchLen)
        t.right(20)
        spring_tree(branchLen-15,t)
        t.left(40)
        spring_tree(branchLen-15,t)
        t.right(20)
        t.pu()
        t.backward(branchLen)
        
        
    if b > 45 and b < 60:
        t.color('brown')
        t.pd()
        t.pensize(3)
        t.forward(branchLen)
        t.right(20)
        spring_tree(branchLen-15,t)
        t.left(40)
        spring_tree(branchLen-15,t)
        t.right(20)
        t.pu()
        t.backward(branchLen)
        
        
    if b > 30 and b < 45:
        t.color('brown')
        t.pd()
        t.pensize(2)
        t.forward(branchLen)
        t.right(20)
        spring_tree(branchLen-15,t)
        t.left(40)
        spring_tree(branchLen-15,t)
        t.right(20)
        t.pu()
        t.backward(branchLen)   
        
        
    if b > 5 and b < 30:
        t.pd()
        t.color('green')
        t.pensize(1)
        t.forward(branchLen)
        t.right(20)
        spring_tree(branchLen-15,t)
        t.left(40)
        spring_tree(branchLen-15,t)
        t.right(20)
        t.pu()
        t.backward(branchLen)    
       
        
      
        
        
def spring():
    t = turtle.Turtle()
    t.write("Spring",font=("Arial", 32, "normal"))
    myWin = turtle.Screen()
    t.left(90)
    t.up()
    t.backward(100)
    t.down()
    spring_tree(80,t)
    myWin.exitonclick()
```

```python
import time
import datetime
now = datetime.datetime.now()
month=now.month

m = month

if m < 7 and m > 2:
    spring()
if m < 9 and m > 5:
    summer()
if m < 12 and m > 7:
    autumn()
else:
    winter()
    
```
